```python
data = {}
import json
with open ("/kaggle/input/creating-id/new_data_v1 (1).json",'r') as file:
    data = json.load(file)
    
```

```python
single =[]
for id, value in data.items():
    if id not in single:
        single.append(id)
    for x in data[id]:
        if x not in single:
            single.append(x)
print(len(single))
words = []
```

```python
for x in single:
    #x = x.split('//')[1]
    x = x.split('/')
    for b in x:
        if b not in words:
            words.append(b)
print(len(words))
```

```python
n = len(words)
words_id ={}
id_words = {}
for x in range (0,n):
    string = f"{x:04d}"
    id_words[string] = words[x]
    words_id[words[x]] = string
```
